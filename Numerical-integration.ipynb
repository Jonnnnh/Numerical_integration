### [Метод прямоугольников(средних)](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D1%82%D0%BE%D0%B4_%D0%BF%D1%80%D1%8F%D0%BC%D0%BE%D1%83%D0%B3%D0%BE%D0%BB%D1%8C%D0%BD%D0%B8%D0%BA%D0%BE%D0%B2#:~:text=%D0%9C%D0%B5%D1%82%D0%BE%D0%B4%20%D0%BF%D1%80%D1%8F%D0%BC%D0%BE%D1%83%D0%B3%D0%BE%D0%BB%D1%8C%D0%BD%D0%B8%D0%BA%D0%BE%D0%B2%20%E2%80%94%20%D0%BC%D0%B5%D1%82%D0%BE%D0%B4%20%D1%87%D0%B8%D1%81%D0%BB%D0%B5%D0%BD%D0%BD%D0%BE%D0%B3%D0%BE%20%D0%B8%D0%BD%D1%82%D0%B5%D0%B3%D1%80%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D1%8F,%D0%BA%D0%BE%D0%BD%D1%81%D1%82%D0%B0%D0%BD%D1%82%D1%83%2C%20%D0%BD%D0%B0%20%D0%BA%D0%B0%D0%B6%D0%B4%D0%BE%D0%BC%20%D1%8D%D0%BB%D0%B5%D0%BC%D0%B5%D0%BD%D1%82%D0%B0%D1%80%D0%BD%D0%BE%D0%BC%20%D0%BE%D1%82%D1%80%D0%B5%D0%B7%D0%BA%D0%B5.)

Для приближенного вычисления интеграла используется формула:
$$\int_a^b f(x) \, dx \approx f\left(\frac{a+b}{2}\right) \cdot (b-a)$$
где: 
- $\int_a^b f(x) \, dx$ — определенный интеграл функции $f(x)$ на интервале от $a$ до $b$
- $f\left(\frac{a+b}{2}\right)$ — значение функции в середине интервала интегрирования
- $(b-a)$ — ширина интервала интегрирования, разность между верхним и нижним пределами интегрирования

### [Метод трапеции](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D1%82%D0%BE%D0%B4_%D1%82%D1%80%D0%B0%D0%BF%D0%B5%D1%86%D0%B8%D0%B9#:~:text=%D0%9C%D0%B5%D1%82%D0%BE%D0%B4%20%D1%82%D1%80%D0%B0%D0%BF%D0%B5%D1%86%D0%B8%D0%B9%20%E2%80%94%20%D0%BC%D0%B5%D1%82%D0%BE%D0%B4%20%D1%87%D0%B8%D1%81%D0%BB%D0%B5%D0%BD%D0%BD%D0%BE%D0%B3%D0%BE%20%D0%B8%D0%BD%D1%82%D0%B5%D0%B3%D1%80%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D1%8F,%D0%B3%D1%80%D0%B0%D1%84%D0%B8%D0%BA%D0%BE%D0%BC%20%D1%84%D1%83%D0%BD%D0%BA%D1%86%D0%B8%D0%B8%20%D0%B0%D0%BF%D0%BF%D1%80%D0%BE%D0%BA%D1%81%D0%B8%D0%BC%D0%B8%D1%80%D1%83%D0%B5%D1%82%D1%81%D1%8F%20%D0%BF%D1%80%D1%8F%D0%BC%D0%BE%D1%83%D0%B3%D0%BE%D0%BB%D1%8C%D0%BD%D1%8B%D0%BC%D0%B8%20%D1%82%D1%80%D0%B0%D0%BF%D0%B5%D1%86%D0%B8%D1%8F%D0%BC%D0%B8.)

Формула метода трапеции для численного интегрирования:
$$\int_a^b f(x) \, dx \approx \frac{\Delta x}{2} \left[f(a) + 2 \sum_{i=1}^{n-1} f(x_i) + f(b)\right]$$
где:
- $\int_a^b f(x) \, dx$ — определенный интеграл функции $f(x)$ на интервале от $a$ до $b$
- $\Delta x = \frac{b-a}{n}$ — ширина каждого подынтервала, где \( n \) — количество подынтервалов, на которые разбит исходный интервал
- $f(a)$ и $f(b)$ — значения функции в начальной и конечной точках интервала.
- $\sum_{i=1}^{n-1} f(x_i)$ — сумма значений функции в промежуточных точках $x_i$, где каждая $x_i$ — это точка на интервале, не включая начальную и конечную точки.
- Коэффициент 2 перед суммой отражает то, что все промежуточные точки учитываются дважды, как верхние базы трапеций

### [Метод Симпсона](https://ru.wikipedia.org/wiki/%D0%A4%D0%BE%D1%80%D0%BC%D1%83%D0%BB%D0%B0_%D0%A1%D0%B8%D0%BC%D0%BF%D1%81%D0%BE%D0%BD%D0%B0)

Формула метода Симпсона выражается как:
$$\int_a^b f(x) \, dx \approx \frac{\Delta x}{3} \left[f(x_0) + 4 \sum_{i=1,3,5,\dots}^{n-1} f(x_i) + 2 \sum_{j=2,4,\dots}^{n-2} f(x_j) + f(x_n)\right]$$

где:
- $\int_a^b f(x) \, dx$  — определенный интеграл функции $f(x)$ на интервале от $a$ до $b$
- $\Delta x = \frac{b-a}{n}$ — ширина каждого подинтервала, где n — количество подинтервалов, на которые разбит исходный интервал
- $x_i$ — точки на интервале, в которых вычисляется значение функции $f(x)$, $i = 0, 1, 2, \dots$, $n$
- $f(x_0)$ и $f(x_n)$ — значения функции в начальной и конечной точках интервала
- $\sum_{i=1,3,5,\dots}^{n-1} f(x_i)$ — сумма значений функции в нечётных точках интервала (кроме начальной и конечной), взвешенная весами 4
- $\sum_{j=2,4,\dots}^{n-2} f(x_j)$ — сумма значений функции в чётных точках интервала (исключая начальную и конечную), взвешенная весами 2

## Решение задачи: ##
##### Импорт библиотек и определение функции

```python
%matplotlib widget
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np
from matplotlib import animation

def f(x):
    return np.sin(x)
```

#### Описание: 
- Импортируем необходимые библиотеки для визуализации и работы с массивами.
- Определяем функцию $f(x)$, которую будем интегрировать — в данном случае это синусоида.

##### Установка параметров задачи и создание фигуры

```python
a, b = 0, np.pi
n = 10

x = np.linspace(a, b, 400)
y = f(x)
x_samples = np.linspace(a, b, n + 1)
y_samples = f(x_samples)

fig = plt.figure(figsize=(12, 8))
axes = [fig.add_subplot(1, 3, i+1, projection='3d') for i in range(3)]
```

#### Описание: 
- Задаем границы интегрирования $a$ и $b$, а также количество разбиений n
- Создаем массив $x$ из 400 точек между $a$ и $b$ и вычисляем значения функции $y = f(x)$
- Для методов интегрирования выбираем $n + 1$ равномерно распределенных точек x_samples и вычисляем значения функции в этих точках y_samples
- Создаем фигуру и три подграфика для визуализации трех различных методов численного интегрирования

##### Определение функций для визуализации методов

```python
def plot_method(ax, method='rectangle'):
    if method == 'rectangle':
        for i in range(n):
            ax.bar3d(x_samples[i], 0, 0, (b-a)/n, y_samples[i], 0, color='red', alpha=0.3)
        ax.set_title('Rectangle Method')
    elif method == 'trapezoid':
        for i in range(n):
            verts = [[x_samples[i], 0, 0],
                     [x_samples[i+1], 0, 0],
                     [x_samples[i+1], y_samples[i+1], 0],
                     [x_samples[i], y_samples[i], 0]]
            poly = Poly3DCollection([verts], color='green', alpha=0.3)
            ax.add_collection3d(poly)
        ax.set_title('Trapezoid Method')
    elif method == 'simpson':
        for i in range(n//2):
            idx = 2*i
            xx = x_samples[idx:idx+3]
            yy = y_samples[idx:idx+3]
            zz = np.zeros_like(xx)
            poly = np.polyfit(xx, yy, 2)
            x_poly = np.linspace(xx[0], xx[2], 100)
            y_poly = np.polyval(poly, x_poly)
            z_poly = np.zeros_like(x_poly)
            verts = [list(zip(x_poly, y_poly, z_poly))]
            ax.add_collection3d(Poly3DCollection(verts, facecolors='blue', edgecolors='b', alpha=0.3))
        ax.set_title('Simpson Method')
    
    ax.plot(x, y, np.zeros_like(x), 'k', label='f(x) = sin(x)')
    ax.legend()
    ax.set_zlim(0, 1)
```

#### Описание: 
- Определяем функцию $plot$_$method$, которая рисует график в зависимости от выбранного метода интегрирования $(rectangle, trapezoid, simpson)$
- Для каждого метода создаем соответствующие геометрические формы (бары, треугольники, параболы), используя трехмерные визуализации

##### Создание анимации трехмерных графиков

```python
for ax, method in zip(axes, ['rectangle', 'trapezoid', 'simpson']):
    plot_method(ax, method)

def animate(i):
    for ax in axes:
        ax.view_init(elev=30, azim=i)
    return fig,

anim = animation.FuncAnimation(fig, animate, frames=np.arange(0, 360, 2), interval=50)

plt.show()
```
